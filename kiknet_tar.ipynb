```python
import os
import tarfile, gzip

filePath = 'G:\kik-net-Mj6+'
events = os.listdir(filePath)
```

```python
for event in events:
    print(event)
    files = os.listdir(os.path.join(filePath, event))
    for file in files:
        if file.split('.')[2] != "img":
            tar = tarfile.open(os.path.join(filePath, event, file))  
            names = tar.getnames() 
            for name in names:
                if name.split('.')[-2] != "ps":
#                     print(name.split('.')[-2])
                    tar.extract(name, os.path.join(filePath, event))  
            tar.close() 
```

```python
for event in events:
    print(event)
    files = os.listdir(os.path.join(filePath, event))
    for file in files:
        if file.split('.')[-1] == 'gz':
            os.remove(os.path.join(filePath, event, file))
    
```
